***import lib pandas***

```python
import pandas as pd
```

***Load data CSV***

```python
df = pd.read_csv("/kaggle/input/data-excel/Salary_Data.csv")
```

***Melihat struktur data secara langsung***

- Menampilkan 5 data teratas

```python
df.head()
```

***Melihat deskripsi singkat data yang dimiliki***

* range index = 30 keseluruan dari 0-29
* jumlah kolom = 2
* Non-null = angka

```python
df.info()
```

***Filtering***

* Membuat x, y
* [:, :-1] Memasukkan semua baris, semua kolom kecuali kolom terakhir (Kolom year)
* [:, 1] Semua baris, kolom terakhir (Salary)

```python
x = df.iloc[:, :-1]
y = df.iloc[:, 1]
```

***Memisahkan data menjadi dua bagian***

* (train dan test)
* 1/3 = 30% dari datanya akan dipakai untuk testing
* 70% sisanya untuk train data
* random state 0 = data train sama test itu datanya tetap

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

***Memasukkan Model***

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

***Data yang akan diprediksi***

diprediksi menggunakan x_test

```python
y_pred = regressor.predict(x_test)
```

***Memvisualisasi memakai matplotlib***

* Yang pertama grafik buat training
* Yang kedua grafik buat testing

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, color="green")
plt.plot(x_train, regressor.predict(x_train), color="red")
plt.title("Years Experience VS Salary")
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test, y_test, color="green")
plt.plot(x_train, regressor.predict(x_train), color="red")
plt.title("Years Experience VS Salary")
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.show()
```

***Memprediksi angka baru***

```python
salary_pred = regressor.predict([[12]]) # prediksi angka tahun
print("The salary for that amount of years experience is: ", salary_pred)
```
